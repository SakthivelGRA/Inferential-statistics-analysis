## 2.1 What proportion of the gunny bags have a breaking strength of less than 3.17 kg per sq cm?

```python
from scipy.stats import norm
```

```python
# Given data in the problem statement:
mean = 5  
std_dev = 1.5  
breaking_strength_threshold = 3.17  

# Calculating the proportion using the cumulative distribution function (CDF):
proportion_less_than_3_17 = norm.cdf(breaking_strength_threshold, loc=mean, scale=std_dev)

# Visualizing  the probability using a cumulative distribution plot
plt.figure(figsize=(6, 4))
x_values = np.linspace(mean - 4 * std_dev, mean + 4 * std_dev, 100)
plt.plot(x_values, norm.cdf(x_values, loc=mean, scale=std_dev), color='blue')
plt.fill_between(x_values, norm.cdf(x_values, loc=mean, scale=std_dev), color='lightblue', alpha=0.5, where=(x_values <= breaking_strength_threshold))
plt.xlabel('Breaking Strength (kg per sq. centimeter)')
plt.ylabel('Cumulative Probability')
plt.title('CDF: Breaking Strength of Gunny Bags')
plt.axvline(breaking_strength_threshold, color='red', linestyle='--', label=f'Breaking Strength = {breaking_strength_threshold} kg')
plt.legend()
plt.show()

print("Proportion of gunny bags with a breaking strength of less than 3.17 kg per sq cm:", proportion_less_than_3_17)
```

## 2.2 What proportion of the gunny bags have a breaking strength of at least 3.6 kg per sq cm.?

```python
# Given data in the problem statement:
mean = 5 
std_dev = 1.5 
breaking_strength_threshold = 3.6  

# Calculating the  proportion using the cumulative distribution function (CDF):
proportion_at_least_3_6 = 1 - norm.cdf(breaking_strength_threshold, loc=mean, scale=std_dev)

# Visualizing  the probability using a cumulative distribution plot:
plt.figure(figsize=(6, 4))
x_values = np.linspace(mean - 4 * std_dev, mean + 4 * std_dev, 100)
plt.plot(x_values, norm.cdf(x_values, loc=mean, scale=std_dev), color='blue')
plt.fill_between(x_values, norm.cdf(x_values, loc=mean, scale=std_dev), color='lightblue', alpha=0.5, where=(x_values >= breaking_strength_threshold))
plt.xlabel('Breaking Strength (kg per sq. centimeter)')
plt.ylabel('Cumulative Probability')
plt.title('CDF: Breaking Strength of Gunny Bags')
plt.axvline(breaking_strength_threshold, color='red', linestyle='--', label=f'Breaking Strength = {breaking_strength_threshold} kg')
plt.legend()
plt.show()

print("Proportion of gunny bags with a breaking strength of at least 3.6 kg per sq cm:", proportion_at_least_3_6)
```

## 2.3 What proportion of the gunny bags have a breaking strength between 5 and 5.5 kg per sq cm.?

```python
# Given data in the problem statement:
mean = 5  
std_dev = 1.5 
lower_bound = 5  
upper_bound = 5.5  

# Calculating the probabilities using the cumulative distribution function (CDF):
prob_lower_bound = norm.cdf(lower_bound, loc=mean, scale=std_dev)
prob_upper_bound = norm.cdf(upper_bound, loc=mean, scale=std_dev)

# Calculate proportion between the two values
proportion_between = prob_upper_bound - prob_lower_bound

# Visualizing the probability using a cumulative distribution plot:
plt.figure(figsize=(6, 4))
x_values = np.linspace(mean - 4 * std_dev, mean + 4 * std_dev, 100)
plt.plot(x_values, norm.cdf(x_values, loc=mean, scale=std_dev), color='blue')
plt.fill_between(x_values, norm.cdf(x_values, loc=mean, scale=std_dev), color='lightblue', alpha=0.5, where=(x_values >= lower_bound) & (x_values <= upper_bound))
plt.xlabel('Breaking Strength (kg per sq. centimeter)')
plt.ylabel('Cumulative Probability')
plt.title('CDF: Breaking Strength of Gunny Bags')
plt.axvline(lower_bound, color='red', linestyle='--', label=f'Lower Bound = {lower_bound} kg')
plt.axvline(upper_bound, color='green', linestyle='--', label=f'Upper Bound = {upper_bound} kg')
plt.legend()
plt.show()

print("Proportion of gunny bags with breaking strength between 5 and 5.5 kg per sq cm:", proportion_between)
```

## 2.4 What proportion of the gunny bags have a breaking strength NOT between 3 and 7.5 kg per sq cm.?

```python
# Given data in the problem statement:
mean = 5 
std_dev = 1.5  
lower_bound = 3  
upper_bound = 7.5  

# Calculate probabilities using the cumulative distribution function (CDF):
prob_lower_bound = norm.cdf(lower_bound, loc=mean, scale=std_dev)
prob_upper_bound = norm.cdf(upper_bound, loc=mean, scale=std_dev)

# Calculating the proportion of gunny bags not between 3 and 7.5 kg per sq cm:
proportion_between = 1 - (prob_upper_bound - prob_lower_bound)

# Visualizing the probability using a cumulative distribution plot:
plt.figure(figsize=(6, 4))
x_values = np.linspace(mean - 4 * std_dev, mean + 4 * std_dev, 100)
plt.plot(x_values, norm.cdf(x_values, loc=mean, scale=std_dev), color='blue')
plt.fill_between(x_values, norm.cdf(x_values, loc=mean, scale=std_dev), color='lightblue', alpha=0.5, where=((x_values < lower_bound) | (x_values > upper_bound)))
plt.xlabel('Breaking Strength (kg per sq. centimeter)')
plt.ylabel('Cumulative Probability')
plt.title('CDF: Breaking Strength of Gunny Bags')
plt.axvline(lower_bound, color='red', linestyle='--', label=f'Lower Bound = {lower_bound} kg')
plt.axvline(upper_bound, color='green', linestyle='--', label=f'Upper Bound = {upper_bound} kg')
plt.legend()
plt.show()

print("Proportion of gunny bags with breaking strength NOT between 3 and 7.5 kg per sq cm:", proportion_between)
```

###  Null Hypothesis :

* The mean Brinell's hardness index of unpolished stones is equal to or greater than 150 (suitable for printing).

### Alternative Hypothesis:

* The mean Brinell's hardness index of unpolished stones is less than 150 (not suitable for printing).

```python
unpolished_hardness = df1['Unpolished ']

# Performing the sample test:
test_statistic, p_value = stats.ttest_1samp(unpolished_hardness, 150)

# Given significance level
alpha = 0.05
print("Test Statistic:", test_statistic)
print("P-value:", p_value)

# Checking the Null hypothesis should be rejected or not:
if p_value < alpha:
    print("Reject the null hypothesis. There is sufficient evidence to conclude that the mean Brinell's hardness index of unpolished stones is less than 150.")
else:
    print("Fail to reject the null hypothesis. There is not enough evidence to conclude that the mean Brinell's hardness index of unpolished stones is less than 150.")
```

## 3.2 Is the mean hardness of the polished and unpolished stones the same?

## Null Hypothesis : 
* The mean hardness of polished stones is equal to the mean hardness of unpolished stones.

## Alternative Hypothesis : 
* The mean hardness of polished stones is not equal to the mean hardness of unpolished stones.

```python
polished_hardness = df1['Treated and Polished']
unpolished_hardness = df1['Unpolished ']

# Performing two sample t-test:
test_statistic, p_value = stats.ttest_ind(polished_hardness, unpolished_hardness)

# Given significance level:
alpha = 0.05
print("Test Statistic:", test_statistic)
print("P-value:", p_value)

# Checking that the null hypothesis can be rejected or not:
if p_value < alpha:
    print("Reject the null hypothesis. There is sufficient evidence to conclude that the mean hardness of polished and unpolished stones is not the same.")
else:
    print("Fail to reject the null hypothesis. There is not enough evidence to conclude that the mean hardness of polished and unpolished stones is not the same.")
```

```python
# Separating the data into groups based on dentist:
dentist_groups = [group['Response'] for name, group in df2.groupby('Dentist')]

#Checking the  Normality assumption:
normality_results = [shapiro(group) for group in dentist_groups]
print("Shapiro-Wilk Normality Test Results:")
for dentist, result in zip(df2['Dentist'].unique(), normality_results):
    print(f"Dentist {dentist}: p-value = {result[1]}")

# Checking if there are at least two groups for Levene's test:
if len(dentist_groups) < 2:
    print("Error: At least two dentist groups with sufficient data are required for Homogeneity of Variances test.")
else:
    # Checking the Homogeneity of Variances assumption:
    homogeneity_result = levene(*dentist_groups)
    print("\nLevene's Test for Homogeneity of Variances:")
    print("Test Statistic:", homogeneity_result[0])
    print("P-value:", homogeneity_result[1])
    
for dentist, group in df2.groupby('Dentist'):
        plt.figure(figsize=(8, 6))
        plt.title("Residuals for Dentist {dentist}")
        plt.hist(group['Response'] - group['Response'].mean(), bins=10)
        plt.xlabel("Residuals")
        plt.ylabel("Frequency")
        plt.show()
```

```python
from scipy.stats import f_oneway

# Checking  if there are at least two groups with sufficient data for the ANOVA test
if len(dentist_groups) < 2:
    print("Error: At least two dentist groups with sufficient data are required for the ANOVA test.")
else:
    # Perform the one-way ANOVA test
    f_statistic, p_value = f_oneway(*dentist_groups)

    # Print the results
    print("One-way ANOVA Test Results:")
    print("F-statistic:", f_statistic)
    print("P-value:", p_value)

# Given significance level:
    alpha = 0.05
    if p_value < alpha:
        print("The p-value is less than the significance level, indicating significant differences exist among dentists in terms of implant hardness.")
    else:
        print("The p-value is greater than or equal to the significance level, indicating no significant differences exist among dentists in terms of implant hardness.")
```

## Hypothesis Test:

One-way ANOVA Test: The p-value obtained is approximately 0.112, which is greater than the significance level of 0.05. Therefore, we fail to reject the null hypothesis. This indicates that there are no significant differences in implant hardness among the different dentists.

## Conclusion:
Based on the results of the hypothesis test, we conclude that there are no significant differences in implant hardness among the different dentists.

## 4.2 How does the hardness of implants vary depending on methods?

### Null Hypothesis : 
 * The mean hardness of implants is the same across all methods.
 
### Alternative Hypothesis : 
* The mean hardness of implants varies depending on the method used.

```python
# Loding the data set:
df3 = pd.read_excel("Dental+Hardness+data.xlsx")

# Grouping the given data by methods:

method_groups = [group['Response'] for name, group in df3.groupby('Method')]

#Checking the  normality of residuals within each method group using Shapiro-Wilk test:
normality_results = {}
for method, data in zip(df3['Method'].unique(), method_groups):
    shapiro_stat, shapiro_p_value = shapiro(data)
    normality_results[method] = {'Shapiro-Wilk Statistic': shapiro_stat, 'P-value': shapiro_p_value}

print("Shapiro-Wilk Normality Test Results:")
for method, result in normality_results.items():
    print(f"Method {method}: Shapiro-Wilk Statistic = {result['Shapiro-Wilk Statistic']}, P-value = {result['P-value']}")

#checking the homogeneity of variances across groups using Levene's test:
levene_stat, levene_p_value = levene(*method_groups)
print("\nLevene's Test for Homogeneity of Variances:")
print("Test Statistic:", levene_stat)
print("P-value:", levene_p_value)
```

```python
plt.figure(figsize=(8, 4))
sns.boxplot(x='Method', y='Response', data=df2)
plt.title('Variation in Implant Hardness Across Methods')
plt.xlabel('Method')
plt.ylabel('Implant Hardness')
plt.show()
```

```python
# Performing the Welch's ANOVA test:
f_statistic, p_value = f_oneway(*method_groups)
print("Welch's ANOVA Test Results:")
print("F-statistic:", f_statistic)
print("P-value:", p_value)
```

## Conclusion:

### P value : 7.683891892977992e-08

* ANOVA Test shows that there are statistically significant differences in hardness amonng the different methods of dental implantation.Atleast one pair of methods had significantly diffrent hardness.

* Since the p-value is negligible we can reject the Null hypothesis.

```python
import statsmodels.api as sm
from statsmodels.formula.api import ols
model = ols('Response ~ C(Dentist) + C(Method) + C(Dentist):C(Method)', data=df4).fit()
anova_table = sm.stats.anova_lm(model, typ=2)
p_value = anova_table['PR(>F)']['C(Dentist):C(Method)']
print("Two-way ANOVA Test Results:")
print("P-value:", p_value)
```

* The p-value obtained from the two-way ANOVA test is 0.00197. This p-value indicates that there is a statistically significant interaction effect between Dentist and Method on the hardness of dental implants.

* The significance of this result suggests that the effect of one factor (Dentist) on implant hardness depends on the level of the other factor (Method), and vice versa.

```python
from statsmodels.stats.multicomp import pairwise_tukeyhsd

#Defining the function for conducting post-hoc tests:
def conduct_posthoc_tests(data, factor):
    tukey = pairwise_tukeyhsd(endog=data['Response'], groups=data[factor], alpha=0.05)
    print(tukey)
# For Alloy type1:
print("Post-hoc Tests for Alloy Type 1:")
# Post-hoc tests for Dentist factor:
print("\nDentist:")
conduct_posthoc_tests(df4[df4['Alloy'] == 1], 'Dentist')
# Post-hoc tests for Method factor:
print("\nMethod:")
conduct_posthoc_tests(df4[df4['Alloy'] == 1], 'Method')

# For Alloy type2:
print("\nPost-hoc Tests for Alloy Type 2:")
# Post-hoc tests for Dentist factor:
print("\nDentist:")
conduct_posthoc_tests(df4[df4['Alloy'] == 2], 'Dentist')

# Post-hoc tests for Method factor:
print("\nMethod:")
conduct_posthoc_tests(df4[df4['Alloy'] == 2], 'Method')
```
